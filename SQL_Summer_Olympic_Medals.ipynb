### Ranking the countries with the most medals

```python
# Let's check our results
query_rank = '''
SELECT Country_Name, SUM(Total_Medals) AS Performance
FROM olympic_medals
GROUP BY Country_Name
HAVING Performance >= 70
ORDER BY Performance DESC
LIMIT 50;
'''
pd.read_sql_query(query_rank, conn)
# The US seems to dominate with the Soviet Union and other European countries following from afar
```

### Ranking the countries with the most medals after the creation and before the dissolution of the Soviet Union

```python
# Let's check our results
query_rank_adjusted = '''
SELECT Country_Name, SUM(Total_Medals) AS Performance
FROM olympic_medals
WHERE Year BETWEEN 1952 AND 1991 
GROUP BY Country_Name
ORDER BY Performance DESC
LIMIT 50;
'''
pd.read_sql_query(query_rank_adjusted, conn)
# The Soviet Union and the US seem to be at the top, with East Germany being third 
# and other European countries following the formerly communist Germany close 
```

```python
query_medals_per_game = '''
SELECT Country_Name, SUM(total_medals) AS medals, COUNT(total_medals) AS games,
SUM(total_medals)/COUNT(total_medals) AS medals_per_game
FROM olympic_medals
GROUP BY Country_Name
ORDER BY medals_per_game DESC
LIMIT 50;
'''
pd.read_sql_query(query_medals_per_game, conn)
# Overall the Unified Team (Team of former Soviet republics compeeting 1 year after the collapse of the USSR) has the best performance
# Other communist/formerly communist countries also have a high degree of medals per game
# The US is third with other European countries following it
# ROC (Russian athletes competing in 2022) also have a high degree of medals per game
```

### Medals obtained while hosting

```python
query_medals_per_host_country ='''
SELECT 
Host_Country, SUM(total_medals) AS medals, SUM(Gold) AS Gold,
SUM(Silver) AS Silver, SUM(Bronze) AS Bronze,
COUNT(*) AS total_hosting,
28 AS total_olympics
FROM olympic_medals
WHERE Host_Country = Country_Name
GROUP BY Host_Country
ORDER BY total_hosting DESC
LIMIT 50;
'''
pd.read_sql_query(query_medals_per_host_country, conn)
```

### Comparing medals obtained when hosting/not-hosting

```python
query_medals_when_hosting_or_not = '''
SELECT Host_Country, 
'Hosting' AS Type, 
COUNT(*) AS total_hosting,
28 AS total_olympics, 
SUM(total_medals) AS medals, 
SUM(Gold) AS Gold,
SUM(Silver) AS Silver, 
SUM(Bronze) AS Bronze
FROM olympic_medals
WHERE Host_Country = Country_Name
GROUP BY Host_Country

UNION

SELECT Country_Name, 
'Not Hosting' AS Type,
'-' AS total_hosting,
28 AS total_olympics,
SUM(total_medals) AS medals,
SUM(Gold) AS Gold,
SUM(Silver) AS Silver,
SUM(Bronze) AS Bronze
FROM olympic_medals
WHERE Country_Name IN (SELECT DISTINCT Host_Country FROM olympic_medals) 
AND Host_Country <> Country_Name
GROUP BY Country_Name
ORDER BY Host_Country ASC, Type ASC;
'''
pd.read_sql_query(query_medals_when_hosting_or_not, conn)
# As we can see most countries do significantly better when hosting sometimes obtaining more than 20% of their total medals when hosting
# Greece has even obtained more medals when hosting compared to when not!!!
```
